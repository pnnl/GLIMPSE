```python
import bs4 as bs
```

```python
with open("../data/CIM/IEEE123.xml", "r") as xml_file:
   xml_data = xml_file.read()
```

```python
soup = bs.BeautifulSoup(xml_data, "xml")
```

```python
print(f"number of Terminals: {len(soup.find_all('cim:Terminal'))}")
print(f"number of Connectivity Nodes: {len(soup.find_all('cim:ConnectivityNode'))}")
print(f"number of ACLineSegment: {len(soup.find_all('cim:ACLineSegment'))}")
```
